## 1. Overview of the Models Section

The `models` directory contains all neural network architectures and supporting modules for the MobileFCMViTv3 pipeline. It is responsible for defining, constructing, and managing the deep learning models used for breast cancer detection from ultrasound images. The section is modular, allowing easy integration of new architectures and layers.

## 2. Detailed Explanation of Each File

- **mobilefcmvitv3_model.py**: Main model definition combining FCM clustering and MobileViTv3 backbone.
- **mobilevitv3.py**: Implementation of the MobileViTv3 architecture, including transformer and convolutional blocks.
- **fusion_block.py**: Defines the fusion mechanism for combining FCM and MobileViTv3 features.
- **fcm_encoder.py**: Encodes FCM cluster maps for integration with the neural network.
- **efficientnet.py / densenet.py**: Alternative model architectures for benchmarking.
- **layers.py**: Common layer definitions (convolution, normalization, activation).
- **blocks/**: Contains modular blocks like MobileViT, FCM feature encoder.
- **layers/**: Contains custom layers such as attention, MBConv, transformer blocks.
- **__init__.py**: Makes the directory a Python package.

## 3. File Connections and Workflow

The workflow begins with FCM clustering (fcm_encoder.py), which generates cluster maps from input images. These maps are fused with image features using the fusion block (fusion_block.py). The combined features are processed by the MobileViTv3 backbone (mobilevitv3.py), which uses custom layers and blocks (layers.py, blocks/, layers/) for deep feature extraction. The main model (mobilefcmvitv3_model.py) orchestrates this process, allowing for flexible integration and benchmarking with alternative architectures (efficientnet.py, densenet.py).

## 4. Mathematics Behind FCM (Fuzzy C-Means)

Fuzzy C-Means (FCM) is a clustering algorithm where each data point belongs to clusters with varying degrees of membership.

**Objective Function:**
$$
J = \sum_{i=1}^N \sum_{j=1}^C u_{ij}^m \|x_i - c_j\|^2
$$
Where:
- $N$: Number of data points
- $C$: Number of clusters
- $u_{ij}$: Membership of $x_i$ in cluster $j$
- $m$: Fuzziness parameter ($m > 1$)
- $x_i$: Data point
- $c_j$: Cluster center

**Update Equations:**
- Membership:
$$
u_{ij} = \frac{1}{\sum_{k=1}^C \left(\frac{\|x_i - c_j\|}{\|x_i - c_k\|}\right)^{\frac{2}{m-1}}}
$$
- Cluster Centers:
$$
c_j = \frac{\sum_{i=1}^N u_{ij}^m x_i}{\sum_{i=1}^N u_{ij}^m}
$$

**Process:**
1. Initialize cluster centers and memberships
2. Update memberships and centers iteratively
3. Stop when convergence is reached

## 5. Mathematics Behind MobileViTv3

MobileViTv3 combines convolutional neural networks (CNNs) and transformer blocks for efficient image feature extraction.

**Convolutional Layers:**
- Apply filters to input images to extract spatial features.
- Mathematical operation:
$$
Y_{i,j} = \sum_{m=1}^M \sum_{n=1}^N X_{i+m, j+n} \cdot W_{m,n}
$$
Where $X$ is the input, $W$ is the filter, $Y$ is the output.

**Transformer Blocks:**
- Use self-attention to capture global dependencies.
- Self-attention equation:
$$
\text{Attention}(Q, K, V) = \text{softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right)V
$$
Where $Q$ (query), $K$ (key), $V$ (value) are projections of the input, $d_k$ is the dimension.

**MobileViT Block:**
- Combines local (CNN) and global (transformer) features.
- Efficient for mobile devices due to lightweight design.

## 6. Fusion Methodology

Fusion combines FCM cluster maps with image features extracted by MobileViTv3. The fusion block aligns and concatenates cluster features with deep features, enabling the model to leverage both spatial segmentation and learned representations.

**Algorithmic Approach:**
- Cluster maps are encoded and resized to match feature map dimensions.
- Features are concatenated or added channel-wise.
- Fusion block processes combined features before passing to subsequent layers.

**Mathematical Representation:**
$$
F_{fusion} = \text{Concat}(F_{fcm}, F_{cnn})
$$
Where $F_{fcm}$ is the FCM feature map, $F_{cnn}$ is the CNN feature map.

## 7. Layer-by-Layer Explanation of MobileViTv3

- **Input Layer:** Receives preprocessed images and fused cluster maps.
- **Conv-BN-Act Layer:** Convolution, batch normalization, and activation (ReLU/SiLU).
- **MBConv Block:** Mobile inverted bottleneck convolution for efficient feature extraction.
- **MobileViT Block:** Combines convolutional and transformer layers for local and global feature learning.
- **Attention Fusion Layer:** Applies channel or spatial attention to enhance important features.
- **Transformer Block:** Multi-head self-attention and feed-forward layers for global context.
- **Fusion Block:** Concatenates FCM and CNN features.
- **Output Layer:** Fully connected layer for classification (malignant/benign).

Each layer is designed for efficiency and accuracy, supporting mobile deployment.

## 8. Training Process

- **Data Loading:** Images and cluster maps loaded and preprocessed.
- **Preprocessing:** Denoising, normalization, resizing, enhancement.
- **Model Initialization:** MobileFCMViTv3 model constructed with config parameters.
- **Optimizer:** Adam, SGD, or RMSprop used for parameter updates.
- **Loss Function:** Cross-entropy for classification; optionally focal loss for imbalanced data.
- **Training Loop:**
  - Forward pass: Input processed through model
  - Loss computed and backpropagated
  - Parameters updated
  - Metrics logged
- **Checkpointing:** Model weights saved periodically
- **Experiment Tracking:** Metrics and losses logged to tracking platforms (Weights & Biases, MLflow, Neptune)

## 9. Metrics Used for Evaluation

- **Accuracy:** $\frac{TP + TN}{TP + TN + FP + FN}$
- **Precision:** $\frac{TP}{TP + FP}$
- **Recall (Sensitivity):** $\frac{TP}{TP + FN}$
- **F1 Score:** $2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}$
- **Specificity:** $\frac{TN}{TN + FP}$
- **ROC-AUC:** Area under the ROC curve
- **Confusion Matrix:** Visualizes true/false positives/negatives
- **Grad-CAM:** Visual explanation of model predictions

These metrics are computed during evaluation and logged for experiment tracking.
